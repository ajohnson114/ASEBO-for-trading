```python
#Cross sectional momentum
import pandas as pd
import yfinance as yf
import datetime as dt
import pandas_datareader as reader
import numpy as np
from dateutil import relativedelta
```

Look over performance - buy the winners and short the losers

```python
#Get tickers for S&P
tickers = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
tickers = tickers.Symbol.to_list()
tickers = [i.replace(".","-") for i in tickers]
```

```python
tickers
```

```python
#getting prices

start = dt.datetime(2018,1,31)
end = dt.datetime(2020,1,31)

df = reader.get_data_yahoo(tickers[0:30],start,end)['Adj Close']
```

```python
df1 = df["GOOGL"]
```

```python
#env -> row of df new state is next row of df
df1
```

```python
#Calculate monthly returns by asset - sum daily returns

monthly_ret = df.pct_change().resample('M').agg(lambda x:(x+1).prod() - 1)
```

```python
monthly_ret
```

```python
#Calculate returns over past 11 months
past11 = (monthly_ret+1).rolling(11).apply(np.prod)-1
past11
```

```python
port_creation = dt.datetime(2019,12,31)
end_measurement = port_creation - relativedelta(months=1)
```

```python
port_creation
```

```python
end_measurement
```

```python
ret_12 = past11.loc[end_measurement]
```

```python
ret_12 = ret_12.reset_index()
```

```python
ret_12
```

```python
#Creating quitiles for the stocks buy the top quintile short the bottom
ret_12['quintile'] = pd.qcut(ret_12.iloc[:,1],5,labels=False)
```

```python
ret_12
```

```python
winners = ret_12[ret_12.quintile == 4]
losers = ret_12[ret_12.quintile == 0]
```

```python
winners
```

```python
losers
```

```python
winner_ret = monthly_ret.loc[port_creation + relativedelta(months=1),df.columns.isin(winners.Symbols)]
loser_ret = monthly_ret.loc[port_creation + relativedelta(months=1),df.columns.isin(losers.Symbols)]
```

```python
winner_ret
```

```python
loser_ret
```

```python
#Subtract because you're shorting
Momentum_profit = winner_ret.mean() - loser_ret.mean()
```

```python
Momentum_profit
```
